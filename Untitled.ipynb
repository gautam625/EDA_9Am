```python
from sentence_transformers import SentenceTransformer
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
model=SentenceTransformer("all-MiniLM-L6-V2")
model
```

```python
sentences=[
    "I like apple.",
    "I enjoy eating apple",
    "I drive a motercycle"]
```

```python
embeddings = model.encode(sentences)
print(embeddings.shape)
print(embeddings)
```

```python
len(model.encode('A'))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
v1=embeddings[0].reshape(-1,1)
v2=embeddings[1].reshape(-1,1)
v3=embeddings[2].reshape(-1,1)
```

```python
cosine_similarity(v1,v2)
```

```python
cosine_similarity(v1,v3)
```

```python
def cosine_sim(a,b):
    return np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b))
```

```python
v1=embeddings[0]
v2=embeddings[1]
v3=embeddings[2]
cosine_sim(v1,v2)
```

```python
cosine_sim(v1,v3)
```
