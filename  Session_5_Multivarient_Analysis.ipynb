```python
label=visa_df["continent"].unique()
certified_count_list=[]
denied_count_list=[]
for i in label:
    con1=visa_df["continent"]==i
    con2=visa_df["case_status"]=='Certified'
    con3=visa_df["case_status"]=='Denied'
    cert_con=con1&con2
    den_con=con1&con3
    certified_count_list.append(len(visa_df[cert_con]))
    denied_count_list.append(len(visa_df[den_con]))

col=sorted(visa_df["case_status"].unique())
pd.DataFrame(zip(certified_count_list,denied_count_list),index=label,columns=col)
```
