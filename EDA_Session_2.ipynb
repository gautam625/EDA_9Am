```python
outliers_list=outliers_df.tolist()
wage_median=visa_df['prevailing_wage'].median()
wage_data=visa_df['prevailing_wage']
new_wage_data=[]

for i in wage_data:
    if i in outliers_list:
        new_wage_data.append(wage_median)
    else:
        new_wage_data.append(i)
```

**Winsoriaztion**
- whatever the outlier less than lb fill with lb
- whatever the outlier greater than ub fill with ub
- in numpy we have method clip

```python
wage_data=visa_df['prevailing_wage'].values
Q1=np.quantile(wage_data,.25)
Q2=np.quantile(wage_data,.50)
Q3=np.quantile(wage_data,.75)

IQR=Q3-Q1
LF=Q1-IQR*(1.5)
UF=Q3+IQR*(1.5)

new_wag=np.clip(wage_data,LF,UF)

plt.figure(figsize=(14,10))
plt.subplot(2,2,1).hist(wage_data)
plt.subplot(2,2,2).hist(new_wag)
plt.subplot(2,2,3).boxplot(wage_data,vert=False)
plt.subplot(2,2,4).boxplot(new_wage_data,vert=False)
plt.show()
```
