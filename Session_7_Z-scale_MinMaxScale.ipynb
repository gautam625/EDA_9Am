```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
visa_df=pd.read_csv(r"C:\Users\gauta\OneDrive\Desktop\Data\Visadataset.csv")
cat=visa_df.select_dtypes(include='object').columns
num=visa_df.select_dtypes(exclude='object').columns
```

```python
# Step1: take the prevailing_wage data
# Step2: calculate mean of the data
# Step3: calculate std of data
# Step4: Nr=mean-std
# Step5: Z_wage=Nr/std
# Step6: creaate new column  visa_df['Z_wage']=z_wage
```

```python
wage_data=visa_df['prevailing_wage']
Mean=visa_df['prevailing_wage'].mean()
std=visa_df['prevailing_wage'].std()
Nr=wage_data-Mean
z_wage=Nr/std
visa_df['Z_wage']=z_wage
visa_df[["prevailing_wage","Z_wage"]]
```

```python
plt.figure(figsize=(12,6))
plt.suptitle("Raw-Data Vs Z-Data")
plt.subplot(2,2,1).hist(visa_df['prevailing_wage'])
plt.title("prevailing_wage")
plt.subplot(2,2,2).hist(z_wage)
plt.title("Z-Score")
plt.subplot(2,2,3).boxplot(visa_df['prevailing_wage'],vert=False)
plt.subplot(2,2,4).boxplot(z_wage,vert=False)
plt.show()
```

# Min-Max-Scaler

```python
# Step1: read data
# Step2: cal min
# Step3: cal max
# Step4: Nr=step1-step2
# Step5: Dr=step2-step3
# Step6: Nr/Dr
```

```python
wage_data=visa_df['prevailing_wage']
Min=wage_data.min()
Max=wage_data.max()
Nr=wage_data-Min
Dr=Max-Min
visa_df['Min_Max_scaler']=Nr/Dr
visa_df[['prevailing_wage','Min_Max_scaler']]
```

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()

wage_mms=mms.fit_transform(visa_df['prevailing_wage'].values.reshape(-1,1))
visa_df['wage_mms']=wage_mms
visa_df[['prevailing_wage','Min_Max_scaler','wage_mms']]
```
